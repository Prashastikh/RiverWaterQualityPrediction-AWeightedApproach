```
import matplotlib.pyplot as plt 
```

```
Models = ['KNN', 'Logistic', 'SVM', 'NB', 'RF']
TrainDataAccuracy = [98.90,91.91,90.30,74.64,99.87]
TestDataAccuracy = [98.17,91.2,88.93,77.86,99.65]
```

```
x2 = Models  
y2 = TrainDataAccuracy
# plotting the line 2 points  
plt.plot(x2, y2, label = "TrainDataAccuracy") 

x3 = Models  
y3 = TestDataAccuracy
# plotting the line 2 points  
plt.plot(x3, y3, label = "TestDataAccuracy") 

# naming the x axis 
plt.xlabel('x - axis') 
# naming the y axis 
plt.ylabel('y - axis') 
# giving a title to my graph 
plt.title('Models Comparison Graph') 
  
# show a legend on the plot 
plt.legend() 
  
# function to show the plot 
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

X = ['KNN', 'Logistic', 'SVM', 'NB', 'RF']

plt.bar(X[0], 98.17, color = 'b', width = 0.25)
plt.bar(X[1], 91.2, color = 'g', width = 0.25)
plt.bar(X[2], 88.93, color = 'r', width = 0.25)
plt.bar(X[3], 77.86, color = 'b', width = 0.25)
plt.bar(X[4], 99.65, color = 'g', width = 0.25)
plt.title("Test Accuracy Comparison")
plt.xlabel('Models')
plt.ylabel('Test Accuracy')
plt.show()
```

```
X = ['KNN', 'Logistic', 'SVM', 'NB', 'RF']

plt.bar(X[0], 98.90, color = 'r', width = 0.25)
plt.bar(X[1], 91.91, color = 'g', width = 0.25)
plt.bar(X[2], 90.30, color = 'b', width = 0.25)
plt.bar(X[3], 74.64, color = 'g', width = 0.25)
plt.bar(X[4], 99.87, color = 'r', width = 0.25)
plt.title("Train Accuracy Comparison")
plt.xlabel('Models')
plt.ylabel('Train Accuracy')
plt.show()
```

```
barWidth = 0.20
plt.figure(figsize=(20,15), dpi=50)

TrainDataAccuracy = [0.9890, 0.9191, 0.9030, 0.7464,0.9987]
TestDataAccuracy = [0.9817,0.912, 0.8893, 0.7786, 0.9965]
Error = [0.12,0.36,0.11,0.84,0.01]
r1 = np.arange(len(TrainDataAccuracy))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

plt.bar(r1, TrainDataAccuracy, color = 'c', width =barWidth, label='Accuracy of Train')
plt.bar(r2, TestDataAccuracy,  color = 'm', width = barWidth, label='Accuracy of Test')
plt.bar(r3, Error,  color = 'y', width = barWidth, label='Error')
plt.title("Model Comparison", fontweight='bold', fontsize=30)

# Add xticks on the middle of the group bars
plt.ylim(0,1)
plt.xticks([r + barWidth for r in range(len(TrainDataAccuracy))], ['KNN', 'Logistic', 'SVM', 'NB', 'RF'])
plt.legend()
plt.tick_params(axis='both', which='major', labelsize=30)
plt.show()
```
